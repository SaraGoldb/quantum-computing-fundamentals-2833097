```python
from qiskit import *
from qiskit.visualization import plot_state_qsphere, array_to_latex
from qiskit.quantum_info import Statevector
%matplotlib inline
```

```python
# Initialize a QuantumCircuit with four qubits
# Configure the qubits into the basis state 1101
circuit = QuantumCircuit(4)
circuit.x([0, 2, 3])
circuit.draw(output='mpl')
```

```python
# UPDATE: Code updated after recording to resolve breaking changes from Qiskit 1.0
circuit.remove_final_measurements()  # no measurements allowed
from qiskit.quantum_info import Statevector
statevector = Statevector(circuit)
plot_state_qsphere(statevector)
```
